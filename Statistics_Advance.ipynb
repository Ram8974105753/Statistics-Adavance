21. What are the assumptions for applying the Central Limit Theorem?
     - Main Assumptions for Applying the CLT:
1. Independent Samples
Each observation in the sample must be independent of the others.

This means the selection of one data point doesn't influence another.

Example: Random sampling from a large population.

If you're sampling without replacement, the sample size should be less than 10% of the population to maintain independence.

2. Identically Distributed
The observations should come from the same population distribution (same mean and variance).

That is, each random variable should have the same probability distribution.

3. Sample Size (n) Is "Large Enough"
The required size depends on the shape of the population distribution:

If population is normal: CLT applies even for small n

If population is not normal: A larger sample size is needed

Rule of thumb:
𝑛
≥
30
n≥30

More skewed or heavy-tailed distributions may need larger samples for the CLT to "kick in"

4. Finite Variance
The population from which the sample is drawn must have a finite variance (i.e., standard deviation must exist and not be infinite).

```python
#3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): Value (0 or 1)
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)
    # Example usage
print("P(X=0):", bernoulli_pdf(0, 0.3))
print("P(X=1):", bernoulli_pdf(1, 0.3))
```

```python
#4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 10000  # number of simulations

# Generate binomial samples
samples = np.random.binomial(n, p, size=size)

# Plot histogram
plt.hist(samples, bins=range(n+2), align='left', edgecolor='black', rwidth=0.8, density=True)
plt.title(f'Binomial Distribution Histogram (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.xticks(range(n+1))
plt.show()
```

```python
#5. Create a Poisson distribution and visualize it using Python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 3.0      # average rate (λ or mu)
size = 10000  # number of samples

# Generate Poisson-distributed samples
data = np.random.poisson(lam=mu, size=size)

# Plot histogram
plt.hist(data, bins=np.arange(0, max(data)+1.5)-0.5, edgecolor='black', color='lightgreen', rwidth=0.85)
plt.title(f"Poisson Distribution (λ = {mu})")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.grid(True)
plt.xticks(np.arange(0, max(data)+1))
plt.show()
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define range for discrete uniform distribution
low = 1     # inclusive
high = 7    # exclusive (1 to 6)

# Create the distribution
dist = randint(low, high)

# Values over which to compute the CDF
x = np.arange(low - 1, high + 1)  # include one value before and after for clarity
cdf_values = dist.cdf(x)

# Plot CDF
plt.step(x, cdf_values, where='post', color='blue', label='CDF')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.xticks(x)
plt.ylim(0, 1.1)
plt.legend()
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for a dataset and plots a histogram.

    Parameters:
    data (array-like): Input numerical dataset

    Returns:
    np.ndarray: Array of Z-scores
    """
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero — Z-scores cannot be computed.")

    z_scores = (data - mean) / std

    # Plotting Z-scores
    plt.hist(z_scores, bins=50, color='mediumseagreen', edgecolor='black', density=True)
    plt.title("Histogram of Z-Scores")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage:
data = np.random.normal(loc=100, scale=15, size=1000)  # Simulated dataset
z_scores = calculate_and_plot_z_scores(data)
```

```python
#18. 5 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a given data point.
    """
    return (x - mean) / std_dev

def plot_z_score(z):
    """
    Plot the standard normal distribution and indicate the Z-score on it.
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.plot(x, y, label="Standard Normal PDF", color="navy")
    plt.fill_between(x, 0, y, where=(x < z), color="lightcoral", alpha=0.5, label="Area ≤ Z")
    plt.axvline(z, color="red", linestyle="--", label=f"Z = {z:.2f}")

    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: Given values
data_point = 85
mean = 70
std_dev = 10

# Step 1: Calculate Z-score
z = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for x = {data_point} (mean = {mean}, std dev = {std_dev}): {z:.2f}")

# Step 2: Visualize it on the standard normal distribution
plot_z_score(z)
```

```python
#19.  Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for a sample mean.

    Parameters:
    - sample: array-like, the sample data
    - population_mean: float, hypothesized population mean (μ₀)
    - population_std: float, known population standard deviation (σ)
    - alpha: float, significance level (default = 0.05)

    Returns:
    - Z-statistic, p-value, and test result
    """
    sample_mean = np.mean(sample)
    n = len(sample)

    # Calculate Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Decision
    reject_null = p_value < alpha

    return z, p_value, reject_null

# Example data
np.random.seed(42)
sample_data = np.random.normal(loc=105, scale=15, size=50)  # Sample from a population with mean=105

# Test against H₀: μ = 100 (null hypothesis)
population_mean = 100
population_std = 15  # Assume population std dev is known

# Perform the Z-test
z_stat, p_val, reject = z_test(sample_data, population_mean, population_std)

# Output results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print("Result:", "Reject H₀" if reject else "Fail to reject H₀")
```

```python
#20.Create a confidence interval for a dataset using Python and interpret the result5
import numpy as np
from scipy import stats

# Example dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=8, size=30)  # Sample of 30 from N(50, 8)

# Step 1: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample std deviation
n = len(data)

# Step 2: Choose confidence level (e.g., 95%)
confidence = 0.95
alpha = 1 - confidence
df = n - 1

# Step 3: Calculate the margin of error using the t-distribution
t_critical = stats.t.ppf(1 - alpha/2, df)
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 4: Compute the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Output the result
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_stats(size=1000, mean=0, std=1, print_result=True):
    """
    Generates a normal distribution dataset and computes skewness and kurtosis.

    Parameters:
    - size (int): number of data points
    - mean (float): mean of the normal distribution
    - std (float): standard deviation of the normal distribution
    - print_result (bool): whether to print results

    Returns:
    - skewness (float)
    - kurtosis (float)
    """
    # Generate the data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    excess_kurtosis = kurtosis(data)  # Fisher definition: normal = 0

    if print_result:
        print(f"Skewness: {skewness:.4f} (0 for perfectly symmetric)")
        print(f"Kurtosis: {excess_kurtosis:.4f} (0 for normal distribution)")

    return skewness, excess_kurtosis

# Run the function
analyze_distribution_stats()
```
